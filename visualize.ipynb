```python
# pandas is for loading and processsing csv file
# matplotlib is a commonly used drawing toolkit
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the csv file
df = pd.read_csv('data.csv')
```

```python
# draw the scatter plot
plt.scatter(df['bmi'], df['target'], alpha=0.5)
plt.title("bmi to target")
plt.xlabel("bmi")
plt.ylabel("target")
plt.show
```

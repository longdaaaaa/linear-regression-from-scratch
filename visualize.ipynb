```python
# pandas is for loading and processsing csv file
# matplotlib is a commonly used drawing toolkit
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the csv file
df = pd.read_csv('data.csv')
```

```python
# draw the scatter plot
plt.scatter(df['AveRooms'], df['MedHouseVal'], alpha=0.5)
plt.title("Average Rooms vs Median House Value (in $1000s)")
plt.xlabel("Average Rooms per Household")
plt.ylabel("Median House Value ($1000s)")
plt.show
```
